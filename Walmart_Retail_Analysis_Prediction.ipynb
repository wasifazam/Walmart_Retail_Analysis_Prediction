```python
Y_target = walmart_dataset['Weekly_Sales']
X_features = walmart_dataset[['Unemployment', 'CPI']]
```

```python
#Split the dataset to test and train the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features, Y_target, test_size=0.20, random_state=0)
```

```python
# Create a linear regression model using the training set
from sklearn.linear_model import LinearRegression
lineReg = LinearRegression()
lineReg.fit(X_train, y_train)
```

```python
y_pred = lineReg.predict(X_test)
y_pred
```

```python
print('Accurate: ', y_test.values[0:30])
print('Predict: ', y_pred[0:30])
```

```python
print('The MSE value is %.2f' %np.mean((lineReg.predict(X_test)-y_test))**2)

from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(lineReg.predict(X_test), y_test)))
```
